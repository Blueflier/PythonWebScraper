```python
import requests
from bs4 import BeautifulSoup

URL = "https://cafebiola.cafebonappetit.com/cafe/cafe-biola/2023-07-14/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
meal_container = soup.find("div",class_="l-container site-panel__daypart-container")

tabs = meal_container.find("div",class_="c-tabs")
tabs_content = tabs.find("div",class_="c-tab__content")
stations = tabs_content.find_all("div", class_="station-title-inline-block")




# job_elements = results.find_all("h3", class_="site-panel__daypart-station-title")
for station in stations:
    station_title = station.find("h3",class_="site-panel__daypart-station-title")
    item_title = station.find("button",class_="site-panel__daypart-item-title")
    item_description = station.find("div",class_="site-panel__daypart-item-description")
    print(station_title.text)
    print(item_title.text)
    if(item_description):
        print(item_description.text)
#breakfast lunch dinner
# meals = soup.find_all("h2", class_="panel__title")

# # print(meals) #this works

# for meal in meals:
#     print(meal.text, end="\n"*2) #this works
    
#     stations = meal.find_all("h3", class_="site-panel__daypart-station-title")
# #     print(stations)
# #     for station in stations:
# #         print(station.text)
```
